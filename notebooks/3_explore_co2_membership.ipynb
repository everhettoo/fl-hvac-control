# CO2 Concentration Fuzzy Membership Function
In this notebook the identified data are used for designing the fuzzy membership functions.

```python
# Imports necessary modules/packages and sets up the environment for experimentation.
import sys
import os

from PIL.ImageColor import colormap

# Navigate up to the project root directory dynamically.
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

# Add the project root to the system path to import custom modules from mylibs package.
sys.path.append(project_root)

# Import membership functions from the mylibs package.
import mylibs.membership_functions as mf
import numpy as np
import matplotlib.pyplot as plt
```

# Define Universe of Discourse and Membership Functions

```python
# Universe of discourse
# Define typical CO2 concentration ranges for indoor air quality analysis (in ppm).
co2_uod = {
    "min": 400,
    "max": 1500,
    "unit": "ppm",
    "linguistic_terms": {
        "Low": {"range": (400, 600)},  # Low occupancy
        "Medium": {"range": (600, 1000)},  # Medium occupancy
        "High": {"range": (1000, 1500)},  # High occupancy
    },
}

"""Input discretized universes"""
co2 = np.linspace(co2_uod["min"], co2_uod["max"], 500)
```

# Generate and Plot Increasing Membership Functions

```python
# Generate increasing membership values
mu_inc_low = [mf.inc(x, 400, 600) for x in co2]
mu_inc_med = [mf.inc(x, 600, 1000) for x in co2]
mu_inc_high = [mf.inc(x, 1000, 1500) for x in co2]
```

```python
# Plot increasing membership functions

plt.figure(figsize=(12, 6))
plt.plot(co2, mu_inc_low, label="Low", color="skyblue")
plt.plot(co2, mu_inc_med, label="Medium", color="green")
plt.plot(co2, mu_inc_high, label="high", color="red")
plt.title("Increasing Fuzzy Membership Functions")
plt.xlabel("CO2")
plt.ylabel("HVAC Membership Degree")
plt.legend()
plt.grid(True)
plt.show()
```

# Generate and Plot Decreasing Membership Functions

```python
# Generate decreasing membership values
mu_dec_low = [mf.dec(x, 400, 600) for x in co2]
mu_dec_med = [mf.dec(x, 600, 1000) for x in co2]
mu_dec_high = [mf.dec(x, 1000, 1500) for x in co2]
```
