# Cooling Level Fuzzy Membership Function
In this notebook the identified data are used for designing the fuzzy membership functions.

```python
# Imports necessary modules/packages and sets up the environment for experimentation.
import sys
import os

# Navigate up to the project root directory dynamically.
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

# Add the project root to the system path to import custom modules from mylibs package.
sys.path.append(project_root)

# Import membership functions from the mylibs package.
from mylibs.increasing_mem import inc
from mylibs.decreasing_mem import dec
from mylibs.triangular_mem import tri
from mylibs.trapeziod_mem import trap
from mylibs.gaussion_mem import gauss
from mylibs.sigmoid_mem import sigmoid
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Define Universe of Discourse and Linguistic Terms

```python
# Universe of discourse
# Define typical CO2 concentration ranges for indoor air quality analysis (in ppm).
cooling_uod = {
    'min': 0,
    'max': 100,
    'unit': '%',
    'linguistic_terms': {
        'OFF': {'range': (0, 25), 'typical_value': 0},
        'LOW': {'range': (20, 40), 'typical_value': 25},
        'MEDIUM': {'range': (40, 70), 'typical_value': 50},
        'HIGH': {'range': (70, 100), 'typical_value': 85}
    }
}

"""Input discretized universes"""
cooling_universes = np.linspace(cooling_uod['min'], cooling_uod['max'], 400)
```

# Generate and Plot Increasing Membership Functions

```python
# Generate increasing membership values
mu_inc_off = [inc(x, 0, 25) for x in cooling_universes]
mu_inc_low = [inc(x, 20, 40) for x in cooling_universes]
mu_inc_med = [inc(x, 40, 70) for x in cooling_universes]
mu_inc_high = [inc(x, 70, 100) for x in cooling_universes]
```

```python
# Plot increasing membership functions

plt.figure(figsize=(10,6))
plt.plot(cooling_universes, mu_inc_off, label='Increasing (0,25)')
plt.plot(cooling_universes, mu_inc_low, label='Increasing (20,40)')
plt.plot(cooling_universes, mu_inc_med, label='Increasing (40,70)')
plt.plot(cooling_universes, mu_inc_high, label='Increasing (70,100)')
plt.title("Increasing (ramp) Fuzzy Membership Functions")
plt.xlabel("Universe of Discourse")
plt.ylabel("Cooling Membership Degree")
plt.legend()
plt.grid(True)
plt.show()
```

# Compare All Membership Functions

```python
df = pd.DataFrame({
    "cooling": cooling_universes,
    "inc_low": mu_inc_low,
    "inc_med": mu_inc_med,
    "inc_high": mu_inc_high,
    "dec_low": mu_dec_low,
    "dec_med": mu_dec_med,
    "dec_high": mu_dec_high,
    "tri_low": mu_tri_low,
    "tri_med": mu_tri_med,
    "tri_high": mu_tri_high,
    "trap_low": mu_trap_low,
    "trap_med": mu_trap_med,
    "trap_high": mu_trap_high,
    "gauss_low": mu_gauss_low,
    "gauss_med": mu_gauss_med,
    "gauss_high": mu_gauss_high,
    "sigmoid_low": mu_sigmoid_low,
    "sigmoid_med": mu_sigmoid_med,
    "sigmoid_high": mu_sigmoid_high
})

def choose_best_two_memberships(df, sets_dict):
    """
    Compare membership functions for each fuzzy set and return the best and second-best.

    Parameters:
        df (pd.DataFrame): DataFrame containing universe + membership values
        sets_dict (dict): dictionary mapping fuzzy set names to column lists

    Returns:
        dict: best and second-best membership function per fuzzy set
    """
    def mse(col1, others):
        avg_other = np.mean([df[c] for c in others], axis=0)
        return np.mean((df[col1] - avg_other)**2)

    best_results = {}

    for set_name, cols in sets_dict.items():
        scores = {}
        for col in cols:
            others = [c for c in cols if c != col]
            scores[col] = mse(col, others)

        # Sort by MSE (lower is better)
        sorted_scores = sorted(scores.items(), key=lambda x: x[1])

        best_results[set_name] = {
            "Best": sorted_scores[0][0],
            "Best MSE": sorted_scores[0][1],
            "Second Best": sorted_scores[1][0],
            "Second Best MSE": sorted_scores[1][1]
        }

    return best_results


# --- Example usage for CO2 fuzzy sets ---
sets_dict = {
    "Low":    ["inc_low","dec_low","tri_low","trap_low","gauss_low","sigmoid_low"],
    "Medium": ["inc_med","dec_med","tri_med","trap_med","gauss_med","sigmoid_med"],
    "High":   ["inc_high","dec_high","tri_high","trap_high","gauss_high","sigmoid_high"]
}

best_two = choose_best_two_memberships(df, sets_dict)
print(best_two)
```
